## Task 1.3 — What the Paper Claims to Improve (9 marks)

### Baseline Identification
**Baseline**: The dual formulation of the Laplacian Support Vector Machine (LapSVM) originally introduced by Belkin, Niyogi, and Sindhwani (2006). (Section 1)

### Limitation of the Baseline
**Limitation**: The dual formulation of LapSVM imposes an $\mathcal{O}(n^3)$ computational complexity for the single step of solving the dual optimization problem, where $n = l + u$ is the total number of training points (labeled + unlabeled). This cubic scaling with respect to the *entire* dataset (not just the labeled set) severely limits its applicability to large-scale semi-supervised learning tasks. Furthermore, because it is solved in the dual space, it is difficult to monitor convergence directly on the classification manifold to implement early stopping. (Section 1 & 2.1)

### How the Proposed Method Overcomes the Limitation
**Improvement**: By replacing the $L_1$ hinge loss with a strictly convex, twice-differentiable $L_2$ squared hinge loss, the authors enable solving the LapSVM directly in the primal space. They propose two primal optimisers: an exact Newton's method that converges in very few iterations, and a Preconditioned Conjugate Gradient (PCG) solver. The PCG solver reduces the computational complexity to $\mathcal{O}(n^2)$ per iteration. Crucially, solving in the primal space allows the algorithm to directly track the classifier's output $f(x)$ at each step, enabling a robust data-based early stopping criterion that halts training once the sign of the predictions on the data manifold stops changing — substantially reducing total execution time. (Section 1, 3.2, & 4)

### Condition Under Which It Might Not Outperform
**When it would fail to outperform**: If the dataset consists almost entirely of severe outliers or label noise, the proposed primal LapSVM might achieve *worse* generalisation performance than the baseline dual LapSVM. This is because the baseline uses the standard $L_1$ hinge loss (which is robust to outliers), whereas the primal method substitutes it with the $L_2$ squared hinge loss to gain differentiability. The $L_2$ loss penalises margin violations quadratically, meaning severe outliers will exponentially skew the decision boundary in the primal method in a way the dual method would resist.
