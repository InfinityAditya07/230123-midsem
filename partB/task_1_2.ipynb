## Task 1.2 — Key Assumptions (8 marks)

### Assumption 1

**Assumption**: The data generating process has support on a low-dimensional Riemannian manifold embedded in the input space, so that labels change smoothly along geodesic distances on that manifold.

**Why the method needs it**: The entire intrinsic regulariser $\gamma_I (\alpha^\top K + \mathbf{1}^\top b) L (K\alpha + \mathbf{1}b)$ is meaningful only if the graph Laplacian $L$ faithfully approximates the Laplace-Beltrami operator on the true manifold. If data does not lie on such a manifold, the penalty term penalises the wrong geometry and can actively harm classification.

**Violation scenario**: High-dimensional data drawn uniformly from a hypercube (no manifold structure) or data whose intrinsic structure is modelled well only in a very high-dimensional space — in either case the kNN graph Laplacian does not converge to any meaningful differential operator, rendering the penalty meaningless or misleading.

**Paper reference**: Section 2 (Manifold Regularisation), first paragraph — "The support of the probability distribution of data is assumed to have the geometric structure of a Riemannian manifold M." Also Equation 1 where $\|f\|_I^2 = f^\top L f$ is derived from this assumption.

### Assumption 2

**Assumption**: The full kernel (Gram) matrix $K$ and the Newton-step Hessian matrix are non-singular (invertible).

**Why the method needs it**: Newton's update (Equation 10) requires computing the inverse of a block Hessian that contains $K$. If $K$ is singular (e.g., when $n > $ number of distinct kernel evaluations, or with a degenerate polynomial kernel), the expansion coefficients $\alpha^*$ are not unique and the Newton step is undefined or numerically unstable.

**Violation scenario**: When two training data points are strictly identical but have different labels, or when using a linear kernel where the number of training points $n$ strictly exceeds the ambient dimension $d$ (making $K$ rank-deficient). In these cases, the Hessian inversion fails.

**Paper reference**: Section 3.1 (Newton's Method), below Equation 10 — "...we have to compute the inverse of the matrix ... assuming the blocks of the matrix are invertible..."

### Assumption 3

**Assumption**: The $L_2$ squared hinge loss is an adequate surrogate for the original $L_1$ hinge loss for classification performance.

**Why the method needs it**: The entire primal framework relies on replacing $L_1$ with $L_2$ to make the objective function twice-differentiable (strictly convex) so that an exact Hessian can be formulated, enabling Newton's method and PCG. The standard $L_1$ hinge loss only permits subgradient descent.

**Violation scenario**: When the dataset contains an extremely high degree of label noise or extreme outliers. The squared hinge loss penalises margin violations quadratically rather than linearly; a single severely mislabeled outlier can disproportionately dominate the gradient and distort the decision boundary far more than it would under the $L_1$ loss.

**Paper reference**: Section 3 (LapSVM Trained in the Primal) — "In this work we substitute the standard hinge loss function of SVM with its squared formulation. This choice ... produces an objective function that is differentiable and strictly convex."
