## Task 1.1 — Core Contribution / Architecture (8 marks)

### Step-by-Step Method Description

1. **Construct the Kernel**: Given $l$ labeled examples and $u$ unlabeled examples (total $n = l + u$), compute the kernel matrix $K \in \mathbb{R}^{n \times n}$ where $K_{ij} = k(x_i, x_j)$. This defines the ambient reproducing kernel Hilbert space $\mathcal{H}_K$. (Section 2, Equation 3)
2. **Build the Graph Laplacian**: Construct a data adjacency graph using a $k$-nearest neighbours approach with Gaussian edge weights $W_{ij}$. Compute the unnormalised graph Laplacian $L = D - W$, where $D$ is the diagonal degree matrix. This encodes the geometry of the data manifold. (Section 2, Equation 1)
3. **Formulate the Primal Objective**: Instead of using the standard $L_1$ hinge loss and deriving the dual, formulate an objective directly in terms of the expansion coefficients $\alpha \in \mathbb{R}^n$ and intercept $b$. This cost function $J(\alpha, b)$ combines three terms: the $L_2$ squared hinge loss on labeled data, the ambient regulariser $\gamma_A \alpha^\top K \alpha$, and the intrinsic manifold regulariser $\gamma_I (\alpha^\top K + \mathbf{1}^\top b) L (K \alpha + \mathbf{1} b)$. The use of the squared hinge loss ($L_2$) is the crucial difference from the original LapSVM as it makes the objective strictly convex and twice-differentiable. (Section 3, Equation 7)
4. **Compute Gradient and Hessian**: Identify the set of error vectors $E$ (labeled points where the margin condition is violated, $1 - y_i f(x_i) > 0$). Compute the exact gradient and a generalised Hessian matrix for the objective function with respect to $z = [b, \alpha^\top]^\top$. (Section 3.1, Equation 9)
5. **Newton's Method Iteration**: Update $z$ iteratively using the Newton step $z^{t} = z^{t-1} - s H^{-1} \nabla_{z}$, where $s$ is the step size. Convergence requires only a few steps, often solved efficiently using Cholesky factorisation with rank-1 updates. (Section 3.1.1, Equation 8)
6. **Preconditioned Conjugate Gradient (PCG) (Optional for large $n$)**: For large datasets where computing $H^{-1}$ is $\mathcal{O}(n^3)$, apply PCG scaled by a diagonal preconditioner, bringing the complexity down to $\mathcal{O}(n^2)$ per iteration. (Section 3.2)
7. **Data-Based Early Stopping**: Monitor the classifier's output $f(x)$ on the unlabeled data manifold. Stop iterations early when the signs of $f(x)$ on the manifold stop changing between iterations, rather than waiting for strict numerical convergence. (Section 4)

### Summary Sentence
**Summary**: The paper bypasses the $\mathcal{O}(n^3)$ dual complexity of semi-supervised Laplacian SVMs by substituting the $L_1$ hinge loss with an $L_2$ squared hinge loss, enabling the primal objective to be solved rapidly using Newton's method or Preconditioned Conjugate Gradient, further accelerated by a data-driven early stopping criterion.
